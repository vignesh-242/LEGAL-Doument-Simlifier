```python

from transformers import T5ForConditionalGeneration, T5Tokenizer
import os

# Load model and tokenizer
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def simplify_clause(text):
    input_text = "simplify: " + text.strip()
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    output_ids = model.generate(input_ids, max_length=100, num_beams=4, early_stopping=True)
    simplified = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return simplified

def main():
    file_path = input("Enter the path to your legal document (.txt): ").strip()

    if not os.path.isfile(file_path):
        print("❌ File not found.")
        return

    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()

    print("\n🔍 Simplifying your document...\n")

    paragraphs = content.strip().split("\n")
    for i, para in enumerate(paragraphs, 1):
        if para.strip():
            simplified = simplify_clause(para)
            print(f"--- Clause {i} ---")
            print(f"Original: {para.strip()}")
            print(f"Simplified: {simplified}\n")

if __name__ == "__main__":
    main()
```
